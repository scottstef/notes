## Concat()
<b>Puts objects together row wise

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randint(25, size=(10,4)))
df
```

```python
#cut into pieces
pieces =[df[:3], df[3:7], df[7:]]
pieces
```

```python
pd.concat(pieces)
```

## Grouping
<b>Splitting</b> the data into groups based on some criteria</br>
<b>Applying</b> a function to each group independantly<br>
<b>Combining</b> the results back into a data structure<br>


```python
df = pd.DataFrame(
     {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randint(25),
        "D": np.random.randint(100),
    }
)
df
```

```python
#Group by a column label, and then selecting the specific columns to get sum of
# Column A has two different values- bar is there 3 times, foo is there 5
df.groupby('A')[['C','D']].sum()
```

<b>MultiIndex- Grouping by multiple column labels<b>

```python
df.groupby(['A', 'B']).sum()
```
