```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
#load the stylesheet 
plt.style.use('fivethirtyeight')
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.show
```

```python

#load the stylesheet 
plt.style.use('fivethirtyeight')
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
fig = plt.gcf()
plt.show

fig.savefig('my_figure.png')
```
